### 1. Fixing missing values

* We have identified **488** missing values in the 'country' column, **16,340** missing values in the 'agent' column, **112,593** missing values in the 'company' column, and **4** missing values in the 'children' column.
* We will fill these missing values using the modes corresponding to the respective hotels.

```python
df.loc[df['hotel'] == 'City Hotel', 'country'].mode()  # checking mode for country
```

```python
# Filling missing values in the 'country' column with 'PRT' when the hotel is a City Hotel.

df.loc[df['hotel'] == 'City Hotel', 'country'] = df.loc[df['hotel'] == 'City Hotel', 'country'].fillna('PRT')
```

```python
df.loc[df['hotel'] == 'Resort Hotel', 'country'].mode()
```

```python
# Filling missing values in the 'country' column with 'PRT' when the hotel is a Resort Hotel.

df.loc[df['hotel'] == 'Resort Hotel', 'country'] = df.loc[df['hotel'] == 'Resort Hotel', 'country'].fillna('PRT')
```

```python
df.loc[df['hotel'] == 'City Hotel', 'agent'].mode()  # checking mode for agent
```

```python
# Filling missing values in the 'agent' column with 9.0 when the hotel is a City Hotel.

df.loc[df['hotel'] == 'City Hotel', 'agent'] = df.loc[df['hotel'] == 'City Hotel', 'agent'].fillna(9.0)
```

```python
df.loc[df['hotel'] == 'Resort Hotel', 'agent'].mode()
```

```python
# Filling missing values in the 'agent' column with 240.0 when the hotel is a Resort Hotel.

df.loc[df['hotel'] == 'Resort Hotel', 'agent'] = df.loc[df['hotel'] == 'Resort Hotel', 'agent'].fillna(240.0)
```

```python
df.loc[df['hotel'] == 'City Hotel', 'company'].mode()  # checking mode for company
```

```python
# Filling missing values in the 'company' column with 40.0 when the hotel is a City Hotel.

df.loc[df['hotel'] == 'City Hotel', 'company'] = df.loc[df['hotel'] == 'City Hotel', 'company'].fillna(40.0)
```

```python
df.loc[df['hotel'] == 'Resort Hotel', 'company'].mode()
```

```python
# Filling missing values in the 'company' column with 223.0 when the hotel is a Resort Hotel.

df.loc[df['hotel'] == 'Resort Hotel', 'company'] = df.loc[df['hotel'] == 'Resort Hotel', 'company'].fillna(223.0)
```

### 3. Dealing with duplicate data

**Note on Duplicate Rows:**

Upon inspection using the `df.duplicated().sum()` function, it was identified that the dataset contains a total of 31,994 rows flagged as duplicates. The presence of these duplicates is attributed to multiple bookings with similar characteristics. These duplicate entries are not exact copies but exhibit variations in certain columns, reflecting distinct instances of bookings with shared attributes.

**Reasons for Retaining Duplicates:**

1. **Valid Variations:** The identified duplicates may represent valid variations in bookings, such as multiple reservations with similar details but different specifics (e.g., different guests, dates).

2. **Lack of Clear Unique Identifier:** The dataset lacks a definitive unique identifier (e.g., booking ID) to unequivocally distinguish between distinct bookings.

3. **Preservation of Data Integrity:** Retaining these rows allows for the preservation of potentially valuable information related to booking patterns, customer behavior, and hotel preferences.

## Customer Type Booking Frequency

```python
df_filtered = df[df['is_canceled'] == 0]

# Creating a pivot table to get counts of customer types for each hotel
pivot_table = df_filtered.pivot_table(index='customer_type', columns='hotel', aggfunc='size', fill_value=0)

# Sorting the rows based on the sum of values in each row (descending order)
pivot_table = pivot_table.loc[pivot_table.sum(axis=1).sort_values(ascending=False).index]

# Creating a heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='Blues',)
plt.title('Distribution of Bookings by Customer Type')
plt.xlabel('')
plt.ylabel('Customer Type')
plt.show()
```

**Insight**

* **Transient** and **Transient-Party** customer types made the maximum bookings at both City and Resort Hotels.

## Top Lead Times for Non-Canceled Bookings

```python
lead_time_city = city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].value_counts().head(10)
lead_time_resort = resort_hotel.loc[resort_hotel['is_canceled'] == 0, 'lead_time'].value_counts().head(10)

# Combining the results into a single DataFrame
lead_time_df = pd.DataFrame({
    'Lead Time': lead_time_city.index,
    'City Hotel': lead_time_city.values,
    'Resort Hotel': lead_time_resort.values
})

# Melting the DataFrame to long format for seaborn
lead_time_df_long = pd.melt(lead_time_df, id_vars='Lead Time', var_name='Hotel', value_name='Count')

lead_time_df = lead_time_df.sort_values(by='City Hotel', ascending=False)
colors = {'City Hotel': 'royalblue', 'Resort Hotel': 'orange'}

# Plotting
plt.figure(figsize=(14, 7))
sns.barplot(data=lead_time_df_long, x='Lead Time', y='Count', hue='Hotel', order=lead_time_df['Lead Time'], palette=colors)
plt.title('Top 10 Lead Time for Non-Canceled Bookings')
plt.xlabel('Lead Time (days)')
plt.ylabel('Count of Bookings')
plt.show()
```

**Insights:**

* Lead times of **0**, **1**, and **2** days are when the maximum number of guests made reservations at both **City** and **Resort** Hotels. This suggests that guests most preferably book their reservations on the **same day** or **1** or **2** days before, resulting in the highest number of non-canceled successful bookings.

### City Hotel Box Plot Summary of Lead Time for Non-Canceled Bookings

```python
city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].describe()  # Statistical Summary of Lead Time
```

```python
# Q1 (First Quartile)

city_hotel_q1 = city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].quantile(.25)
city_hotel_q1
```

```python
# Q2 (Median)

city_hotel_q2 = city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].quantile(.50)
city_hotel_q2
```

```python
# Q3 (Third Quartile)

city_hotel_q3 = city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].quantile(.75)
city_hotel_q3
```

```python
# Range

city_hotel_range = city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].max() - city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].min()
city_hotel_range
```

```python
# Inter Quartile Range (IQR)

city_hotel_iqr = city_hotel_q3 - city_hotel_q1
city_hotel_iqr
```

```python
# Lower Whisker

city_hotel_lower_whisker = max(city_hotel_q1 - (1.5) * city_hotel_iqr, city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].min())
city_hotel_lower_whisker
```

```python
# Upper Whisker

city_hotel_upper_whisker = min(city_hotel_q3 + (1.5) * city_hotel_iqr, city_hotel.loc[city_hotel['is_canceled'] == 0, 'lead_time'].max())
city_hotel_upper_whisker
```

```python
# Outliers

city_hotel_outliers = city_hotel.loc[(city_hotel['is_canceled'] == 0) & (city_hotel['lead_time'] > city_hotel_upper_whisker)]
city_hotel_outliers['lead_time']
```
